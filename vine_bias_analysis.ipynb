```python
import pandas as pd
```

```python
vine_review_df = pd.read_csv('vine_table.csv', sep = ',')
vine_review_df
```

```python
vine_review_votes_df = vine_review_df[vine_review_df['total_votes'] >= 20]
vine_review_votes_df.count()
```

```python
vine_review_votes_calc_df = vine_review_votes_df[vine_review_votes_df['helpful_votes']/vine_review_votes_df['total_votes'] >= 0.5]
vine_review_votes_calc_df
```

```python
vine_review_Y_df = vine_review_votes_calc_df[vine_review_votes_calc_df['vine'] == 'Y']
vine_review_Y_df
```

```python
vine_review_Y = vine_review_Y_df['review_id'].count()
vine_review_Y
```

```python
vine_review_N_df = vine_review_votes_calc_df[vine_review_votes_calc_df['vine'] == 'N']
vine_review_N_df
```

```python
vine_review_N = vine_review_N_df['review_id'].count()
vine_review_N
```

```python
vine_review_Y_fivestar = len(vine_review_Y_df[vine_review_Y_df['star_rating'] == 5.0])
vine_review_Y_fivestar
```

```python
vine_review_N_fivestar = len(vine_review_N_df[vine_review_N_df['star_rating'] == 5.0])
vine_review_N_fivestar
```

```python
fivestar_Y_perc = ((vine_review_Y_fivestar/vine_review_Y) * 100).astype(str) + '%'
fivestar_Y_perc
```

```python
fivestar_N_perc = ((vine_review_N_fivestar/vine_review_N) * 100).astype(str) + '%'
fivestar_N_perc
```
